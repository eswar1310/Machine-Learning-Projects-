```python
import joblib
from colorama import Fore, Style  # Importing Fore for foreground color and Style for text styles

model = joblib.load('decission_tree_classifier.joblib')  # Replace with your actual model file name

def input_with_color(prompt):
    return float(input(Fore.CYAN + prompt + Style.RESET_ALL))

# 2. Input New Values (Form-like)
print(Fore.GREEN + "Input new values for prediction:" + Style.RESET_ALL)
print(ful_mapping.to_string(index=False))
fulfilment = input_with_color("Enter value for fulfilment: ")
print(service_mapping.to_string(index=False))
shipservicelevel = input_with_color("Enter value for shipservicelevel: ")
print(size_mapping.to_string(index=False))
size = input_with_color("Enter value for size: ")
print(cat_mapping.to_string(index=False))
category = input_with_color("Enter value for category: ")
print(state_mapping.to_string(index=False))
shipstate = input_with_color("Enter value for shipstate: ")
b2b = input_with_color("Enter value for b2b: ")
amount = input_with_color("Enter value for amount: ")

# Add more input prompts for other features as needed

# 3. Apply Values to the Model
new_data = [[fulfilment, shipservicelevel, size, amount, shipstate, b2b, category]]  # Adjust as per your model's input format
predictions = model.predict(new_data)

# 4. Display Predicted Output
print(Fore.YELLOW + "\n0  : Rejected\n1  : Not Rejected" + Style.RESET_ALL)
print(Fore.GREEN + "\nPredicted Output:" + Style.RESET_ALL)
print(predictions)

# 5. Repeat and Experiment (Optional)
while input("Do you want to try another prediction? (yes/no): ").lower() == 'yes':
    print(ful_mapping.to_string(index=False))
    fulfilment = input_with_color("Enter value for fulfilment: ")
    print(service_mapping.to_string(index=False))
    shipservicelevel = input_with_color("Enter value for shipservicelevel: ")
    print(size_mapping.to_string(index=False))
    size = input_with_color("Enter value for size: ")
    print(cat_mapping.to_string(index=False))
    category = input_with_color("Enter value for category: ")
    print(state_mapping.to_string(index=False))
    shipstate = input_with_color("Enter value for shipstate: ")
    b2b = input_with_color("Enter value for b2b: ")
    amount = input_with_color("Enter value for amount: ")

    # Repeat for other features as needed
    new_data = [[fulfilment, shipservicelevel, size, amount, shipstate, b2b, category]]  # Adjust as per your model's input format

    predictions = model.predict(new_data)
    print(Fore.YELLOW + "\n0  : Rejected\n1  : Not Rejected" + Style.RESET_ALL)
    print(Fore.GREEN + "\nPredicted Output:" + Style.RESET_ALL)
    print(predictions)

print(Fore.BLUE + "End of interactive demonstration." + Style.RESET_ALL)
```
